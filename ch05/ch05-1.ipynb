```python
noiseimg=gimg.copy()
```

```python
from random import randint
```

```python
noise= 20

whiteMax =noise/2
blackMax = noise
```

```python
noiseimg=gimg.copy()

for h in range(noiseimg.shape[0]):
  for w in range(noiseimg.shape[1]):
     val=randint(0,100)
     if val <whiteMax:
        noiseimg[h,w]=255
     elif  val > blackMax:
        noiseimg[h,w]=0
cv2_imshow(noiseimg)
```

```python
outimg=cv2.medianBlur(noiseimg,3)
cv2_imshow(outimg)
```
