```python
img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/colabIP/Images/lena256.jpg')
Rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gimg=cv2.cvtColor(Rgb_img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gimg)
```

```python
outimg=cv2.GaussianBlur(gimg,(7,7),3)
cv2_imshow(outimg)
```

```python
mean_img=cv2.blur(gimg,(5,5))
cv2_imshow(mean_img)
```

```python
edgeimg=cv2.addWeighted(gimg,1.0,mean_img,-1.0,0)
cv2_imshow(edgeimg)
```

```python
out_img =cv2.addWeighted(gimg,1.0,mean_img,1.0,0)
cv2_imshow(out_img)
```
