```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/colabIP/Images/lena256.jpg')
gimg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
hist =cv2.calcHist([gimg],[0],None,[256],[0,256])
plt.plot(hist)
```

```python
gOut=cv2.equalizeHist(gimg)
cv2_imshow(gOut)
```

```python
hist =cv2.calcHist([gOut],[0],None,[256],[0,256])
plt.plot(hist)
#
print(hist.shape)
```

```python
cunuhist=hist.copy()

for i in range(1,256):
  cunuhist[i]=cunuhist[i-1]+cunuhist[i]
print(cunuhist)
```

```python
equalhist=cunuhist.copy()
numOfPixes=np.int32(gimg.shape[0]*gimg.shape[1])
equalhist=equalhist/numOfPixes
print(equalhist)
```

```python
equalhist =equalhist*255
print(equalhist)
```

```python
gOut=gimg.copy()
for i in range(gimg.shape[0]):
  for j in range(gimg.shape[1]):
    gOut[i,j]=equalhist[gimg[i,j]]
cv2_imshow(gOut)
```
