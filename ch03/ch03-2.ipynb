```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img =cv2.imread('/content/drive/MyDrive/Colab Notebooks/colabIP/Images/girl.jpg')
```

```python
gPlus = np.zeros((img.shape[0],img.shape[1]), dtype=np.ubyte)
```

```python
val=50
for h in range(img.shape[0]):
  for w in range(img.shape[1]):
    plusval = img[h,w]+val
    if(plusval>=255):
      plusval=255
      gPlus[h,w]=plusval
plt.imshow(gPlus,cmap='gray')
```
